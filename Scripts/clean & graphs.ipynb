```python
num_rows_og1, num_cols_og1 = df1.shape
num_rows_og2, num_cols_og2 = df2.shape


print('original rows for math data:', num_rows_og1)
print('original columns for math data:', num_cols_og1)


print('original rows for portuguese data:', num_rows_og2)
print('original columns for portuguese data:', num_cols_og2)
```

```python
#drop unnecessary columns
to_drop = [
    'mother_job',
    'father_job',
    'school_choice_reason',
    'romantic_relationship',
    'guardian'
]
df1.drop(to_drop, inplace=True, axis=1)
df2.drop(to_drop, inplace=True, axis=1)


#rename columns
column_rename = {
    'address_type' : 'geographic_area',
    'travel_time' : 'commute_time',
    'class_failures' : 'num_class_failed',
    'extra_paid_classes' : 'tutor',
    'activities' : 'extracurricular',
    'nursery_school' : 'preschool'
}
df1.rename(columns=column_rename, inplace=True)
df2.rename(columns=column_rename, inplace=True)
```

```python
#sort data based on age ascending
df1 = df1.sort_values(by=['age'])
df2 = df2.sort_values(by=['age'])


#removing all rows where students are going to school at MS
column_school = 'school'
value_school = 'MS'

df1.drop(df1[df1[column_school] == value_school].index, inplace = True)
df2.drop(df2[df2[column_school] == value_school].index, inplace = True)


#fill in blanks with "Unknown"
df1.fillna('Unknown', inplace=True)
df2.fillna('Unknown', inplace=True)


#save changes to new cvs file
df1.to_csv('new_student_math_clean.csv', index=False)
df2.to_csv('new_student_portuguese_clean.csv', index=False)


#print(df1)
#print(df2)
```

```python
#save the new number of rows and columns
num_rows_new1, num_cols_new1 = df1.shape
num_rows_new2, num_cols_new2 = df2.shape


#calculate the number of rows and columns removed
num_rows_removed1 = num_rows_og1 - num_rows_new1
num_cols_removed1 = num_cols_og1 - num_cols_new1

num_rows_removed2 = num_rows_og2 - num_rows_new2
num_cols_removed2 = num_cols_og2 - num_cols_new2
```

```python
#Print the results
print(f"Rows removed for math data: {num_rows_removed1}")
print(f"Columns removed for math: {num_cols_removed1}")
```

```python
#Print the results
print(f"Rows removed for portuguese data: {num_rows_removed2}")
print(f"Columns removed for portuguese: {num_cols_removed2}")
```

```python
m_df = pd.merge(df1, df2[['student_id', 'grade_1', 'grade_2', 'final_grade']], on='student_id', how='left')
m_df.head(4)
```

```python
#rename the new columns in the merged dataset
column_rename2 = {
    'grade_1_x' : 'm_grade1',
    'grade_2_x' : 'm_grade2',
    'final_grade_x' : 'm_final_grade',
    'grade_1_y' : 'p_grade1',
    'grade_2_y' : 'p_grade2',
    'final_grade_y' : 'p_final_grade'
}
m_df.rename(columns=column_rename2, inplace=True)

#save new modified csv file
m_df.to_csv('hs_performance.csv', index=False)

m_df.tail(3)
```

```python
sex_counts = m_df['sex'].value_counts()
if 'F' in sex_counts.index:
    perc_f = (sex_counts['F']/len(m_df)) * 100
    print(f"Percentage of females: {perc_f}")
else:
    print("No info about females in Sex column.")
```

```python
min_p_grade1 = m_df['p_grade1'].min()
count_min_p_grade1 = (m_df['p_grade1'] == min_p_grade1).sum()

print(f"This was the lowest score in first period portuguese grade: {min_p_grade1}")
print(f"{count_min_p_grade1} students  got the lowest grade.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(m_df['p_final_grade'], bins=20, kde=True)
plt.title('Distribution of Portuguese Final Grade')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=m_df['p_final_grade'])
plt.title('Box Plot of Portuguese Final Grade')
plt.show()
```

```python
Q1 = m_df['p_final_grade'].quantile(0.25)
Q3 = m_df['p_final_grade'].quantile(0.75)
IQR = Q3 - Q1

l_bound = Q1 - 1.5 * IQR
u_bound = Q3 + 1.5 * IQR

outliers = (m_df['p_final_grade'] < l_bound) | (m_df['p_final_grade'] > u_bound)
outlier_values = m_df['p_final_grade'][outliers]

print("Outlier values:")
print(outlier_values)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(m_df['absences'], bins=20, kde=True)
plt.title('Distribution of Absences')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=m_df['absences'])
plt.title('Box Plot of Absences')
plt.show()
```

```python
Q1_absences = m_df['absences'].quantile(0.25)
Q3_absences = m_df['absences'].quantile(0.75)
IQR_absences = Q3_absences - Q1_absences

l_bound_absences = Q1_absences - 1.5 * IQR_absences
u_bound_absences = Q3_absences + 1.5 * IQR_absences

outliers_absences = (m_df['absences'] < l_bound_absences) | (m_df['absences'] > u_bound_absences)
outlier_values_absences = m_df['absences'][outliers_absences]

print("Outlier values for Absences:")
print(outlier_values_absences)
```
